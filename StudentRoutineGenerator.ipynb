```python
class StudentRoutineGenerator:

    def __init__(self, name, courses):
        self.name = name
        self.total_courses = courses

        self.routine = {'Sat/Thurs': {}, 'Sun/Tue': {}, 'Mon/Wed': {}}
        self.courses = []
        print(f"Name: {self.name}\nMaximum Number of Courses: {self.total_courses} \nInitial Routine: {self.routine} ")


    def clash_exist(self, day, time):
        for key, val in self.routine.items():
            if key == day:
                for k2, v2 in val.items():
                    if k2 == time:
                        return v2

        return None

    def addCourses(self, *args):
        for tup in args:
            tmp = tup.split("-")
            x = self.clash_exist(tmp[1], tmp[2])


            if x:
                print(f"Can't take {tmp[0]}. It's clashing with your {x}")

            elif tmp[0] in self.courses:
                print(f"You already have {tmp[0]} in your routine")

            else:
                if len(self.courses) == self.total_courses:
                    print(f"You can't take more than {self.total_courses} courses")
                    break
                else:
                    self.courses.append(tmp[0])
                    self.routine[tmp[1]][tmp[2]] = tmp[0]
                    print(f"Successfully added {tmp[0]}!")


    def dropCourse(self, c):
        if c not in self.courses:
            print("No such course in your routine")

        else:
            for key, val in self.routine.items():
                for k2, v2 in val.items():
                    if v2 == c:
                        self.routine[key].pop(k2)
                        self.courses.remove(c)
                        print(f"Successfully dropped {c}")
                        return


    def showRoutine(self):
        print(f"Updated Routine:\n{self.routine}")
        print("Routine Details:")
        for k1,v1 in self.routine.items():
            print(f"{k1}:")
            for k2,v2 in v1.items():
                print(f"{k2} - {v2}")
print('##################################')
st1 = StudentRoutineGenerator('Harry', 4)
print('------------------------------')
st1.addCourses('CSE110-Mon/Wed-12:30', 'MAT110-Mon/Wed-2:00')
st1.addCourses('ENG101-Sun/Tue-12:30', 'CSE110-Mon/Wed-9:30')
st1.addCourses('PHY111-Sun/Tue-12:30')
print('------------------------------')
st1.showRoutine()
print('------------------------------')
st1.dropCourse('CSE110')
st1.dropCourse('PHY112')
print('------------------------------')
st1.showRoutine()
print('##################################')
st2 = StudentRoutineGenerator('John', 3)
print('------------------------------')
st2.addCourses('MAT110-Mon/Wed-8:00')
st2.addCourses('ENG101-Sat/Thurs-12:30', 'CSE110-Sun/Tue-9:30')
st2.addCourses('PHY111-Sun/Tue-12:30')
print('------------------------------')
st2.showRoutine()
```
